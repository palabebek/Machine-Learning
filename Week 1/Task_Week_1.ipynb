# **AutoMobile Dataset**

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.impute import SimpleImputer
import numpy as np

# Muat data
data = pd.read_csv("/content/Automobile.csv")

# Pilih fitur dan target
features = data.drop('symboling', axis=1).select_dtypes(include=np.number) # Hanya fitur numerik
target = data['symboling']

# Impute missing values
imputer = SimpleImputer(strategy='mean')
features_imputed = imputer.fit_transform(features)

# Bagi data menjadi data latih dan data uji
X_train, X_test, y_train, y_test = train_test_split(features_imputed, target, test_size=0.2, random_state=42)

# Inisialisasi dan latih model Linear Regression
model = LinearRegression()
model.fit(X_train, y_train)

# Prediksi pada data uji
y_pred = model.predict(X_test)

# Hitung MSE
mse = mean_squared_error(y_test, y_pred)
print("MSE:", mse)

# Hitung RMSE
rmse = np.sqrt(mse)
print("RMSE:", rmse)

# Hitung R-squared
r2 = r2_score(y_test, y_pred)
print("R-squared:", r2)

# Visualisasi
metrics = ['MSE', 'RMSE', 'R-squared']
values = [mse, rmse, r2]

plt.figure(figsize=(8, 6))
plt.bar(metrics, values, color=['skyblue', 'lightgreen', 'salmon'])
plt.ylabel('Value')
plt.title('Model Evaluation Metrics')
plt.ylim(0, max(values) * 1.1) # Sesuaikan batas y-axis

# Tambahkan nilai di atas bar
for i, value in enumerate(values):
    plt.text(i, value + 0.01 * max(values), f'{value:.3f}', ha='center')

plt.show()
```

**Penjelasan Matriks Evaluasi:**

**Mean Squared Error (MSE)**

*   MSE mengukur rata-rata dari kuadrat perbedaan antara nilai prediksi dan nilai aktual.
*   Rumus: MSE = 1/n Σ (y_actual - y_predicted)^2

*   Semakin kecil nilai MSE, semakin baik modelnya.

**Root Mean Squared Error (RMSE)**
*  RMSE adalah akar kuadrat dari MSE.

*  Rumus: RMSE = √MSE

*  RMSE memberikan interpretasi dalam satuan yang sama dengan target variabel.

*  Semakin kecil nilai RMSE, semakin baik modelnya.

**R-squared (R²)**
*  R-squared mengukur seberapa baik model regresi linier cocok dengan data aktual.

*  Rumus: R² = 1 - (SSres / SStot)

 * SSres adalah jumlah kuadrat residual.
 * SStot adalah jumlah kuadrat total.

*  Nilai R-squared berkisar antara 0 dan 1.
*  Nilai 1 menunjukkan model yang sempurna, sedangkan nilai 0 menunjukkan model yang tidak lebih baik dari rata-rata.

**Akurasi (Accuracy)**
  * Akurasi mengukur seberapa sering model benar dalam memprediksi.
  * Rumus: Akurasi = (TP + TN) / (TP + TN + FP + FN)
  *  TP = True Positive, TN = True Negative, FP = False Positive, FN = False Negative.

**Presisi (Precision)**

  *  Presisi mengukur seberapa sering model benar dalam memprediksi kelas positif.
  *  Rumus: Presisi = TP / (TP + FP)

**Recall (Sensitivity)**

  *  Recall mengukur seberapa sering model dapat menemukan semua contoh positif.
  *  Rumus: Recall = TP / (TP + FN)

**F1-score**

  *  F1-score adalah rata-rata harmonik dari presisi dan recall.
  *  Rumus: F1-score = 2 * (Presisi * Recall) / (Presisi + Recall)
  *  F1-score digunakan ketika ingin menyeimbangkan presisi dan recall.

**AUC (Area Under the Curve)**

  *  AUC mengukur area di bawah kurva ROC.
  *  AUC memberikan gambaran tentang kinerja model dalam membedakan antara kelas positif dan negatif.
  *  Nilai AUC berkisar antara 0 dan 1. Semakin tinggi nilainya, semakin baik modelnya.

**ROC (Receiver Operating Characteristic)**

  *  Kurva ROC memplot True Positive Rate (TPR) terhadap False Positive Rate (FPR) pada berbagai ambang batas klasifikasi.
  *  Kurva ROC memberikan visualisasi tentang trade-off antara TPR dan FPR.
